```python
# Importação das bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Importanto e apresentando o documento

df = pd.read_csv('dataset_datatran2025.csv', sep=';', encoding='latin-1')
display(df)
```

```python
# Configuração de estilo para os gráficos

sns.set_theme(style="whitegrid") # Estética visual dos gráficos
plt.rcParams['figure.figsize'] = (14, 8) # Criar padrão de tamanho para todos os gráficos
```

```python
# Criação do gráfico de Causas por acidente

# Cria um gráfico em branco
plt.figure()

# Selecionando a variável do DF
top_causas = df['causa_acidente'].value_counts().head(10)

# Criando o gráfico com x e y
ax = sns.barplot(x=top_causas.values, y=top_causas.index, palette='Reds_r')

# Adicionando rótulos
for container in ax.containers:
    ax.bar_label(container, padding=3, fontsize=13, fontweight='bold', color="black")

# Customizando títulos, fontes, tamanhos, margens, etc
plt.title('Principais Causas de Sinistros - PRF 2025', fontsize=16, fontweight='bold', color="black", pad=15)
plt.yticks(fontsize=13, fontweight='bold')
plt.xlabel('Quantidade de Acidentes')
plt.xticks([])

plt.savefig('01_top_causas.png', bbox_inches='tight')
plt.show()
```

```python
# Criação do gráfico de Tipos de Acidentes

# Cria um gráfico em branco
plt.figure()

# Selecionando a variável do DF
tipo_acidente = df['tipo_acidente'].value_counts().head(10)

# Criando o gráfico com x e y
ax = sns.barplot(x=tipo_acidente.values, y=tipo_acidente.index, palette='Reds_r')

# Adicionando rótulos
for container in ax.containers:
    ax.bar_label(container, padding=3, fontsize=13, fontweight='bold', color="black")

# Customizando títulos, fontes, tamanhos, margens, etc
plt.title('Principais Tipos de Acidentes - PRF 2025', fontsize=16, fontweight='bold', color="black", pad=15)
plt.yticks(fontsize=13, fontweight='bold')
plt.xlabel('Quantidade de Acidentes')
plt.xticks([])

plt.savefig('02_tipo_acidentes.png', bbox_inches='tight')
plt.show()
```

```python
# Criação do gráfico de Nº de Sinistros por UF

# Cria um gráfico em branco
plt.figure()

# Selecionando a variável do DF
ordem_uf = df['uf'].value_counts().index

# Criando o gráfico com x e y
ax = sns.countplot(data=df, x='uf', order=ordem_uf, palette='Reds_r')

# Adicionando rótulos
for container in ax.containers:
    ax.bar_label(container, padding=3, fontsize=12, fontweight='bold', color="black", rotation=45)

# Customizando títulos, fontes, tamanhos, margens, etc
plt.title('Número de Sinistros por Estado - PRF 2025', fontsize=14, fontweight='bold', color="black", pad=12)
plt.xticks(fontsize=12, fontweight='bold')
plt.xlabel('UF', fontsize=10)
plt.yticks([])
plt.ylabel('Quantidade de Acidentes', fontsize=12)

plt.savefig('03_sinistros_uf.png', bbox_inches='tight')
plt.show()
```

```python
# Criação do gráfico de Nº de Sinistros por BR

# Cria um gráfico em branco
plt.figure()

# Selecionando as variáveis do DF
top_brs = df['br'].value_counts().head(10)

# Criando o gráfico com x e y
ax = sns.barplot(x=top_brs.index, y=top_brs.values, palette='Reds', hue=top_brs.values, legend=False)

# Adicionando rótulos
for container in ax.containers:
    ax.bar_label(container, padding=3, fontsize=12, fontweight='bold', color="black", rotation=45)

# Customizando títulos, fontes, tamanhos, margens, etc
plt.title('Número de Sinistros por BR - PRF 2025', fontsize=14, fontweight='bold', color="black", pad=12)
plt.xticks(fontsize=12, fontweight='bold')
plt.xlabel('Rodovias_BR', fontsize=10)
plt.yticks([])
plt.ylabel('Quantidade de Acidentes', fontsize=12)

plt.savefig('04_sinistros_br.png', bbox_inches='tight')
plt.show()
```

```python
# Criação do gráfico de Acidentes por questões metereológicas

# Cria um gráfico em branco
plt.figure()

# Selecionando a variável do DF
clima = df['condicao_metereologica'].value_counts().head(10)

# Criando o gráfico com x e y
ax = sns.barplot(x=clima.values, y=clima.index, palette='Reds_r')

# Adicionando rótulos
for container in ax.containers:
    ax.bar_label(container, padding=3, fontsize=13, fontweight='bold', color="black")

# Customizando títulos, fontes, tamanhos, margens, etc
plt.title('Acidentes x Condições Metereológicas - PRF 2025', fontsize=16, fontweight='bold', color="black", pad=15)
plt.yticks(fontsize=13, fontweight='bold')
plt.xlabel('Quantidade de Acidentes')
plt.xticks([])

plt.savefig('05_clima.png', bbox_inches='tight')
plt.show()
```

```python
# Criação do gráfico de Nº de Sinistros por Fase do Dia

# Cria um gráfico em branco
plt.figure()

# Selecionando as variáveis do DF
severidade = df.groupby('fase_dia')[['mortos', 'feridos_graves']].sum().sort_values(by='mortos', ascending=False)

# Criando o gráfico com x e y
ax = severidade.plot(kind='bar', stacked=True, color=["#B61F0F", "#ffa513"])

# Adicionando rótulos
for container in ax.containers:
    ax.bar_label(container, padding=3, fontsize=12, fontweight='bold', color="black")

# Customização da legenda
plt.legend(['Mortos', 'Feridos Graves'], title='Gravidade', title_fontsize='10', fontsize='10')

# Customizando títulos, fontes, tamanhos, margens, etc
plt.title('Número Sinistros por Fase do Dia - PRF 2025', fontsize=14, fontweight='bold', color="black", pad=20)
plt.xticks(fontsize=12, fontweight='bold', rotation=0)
plt.xlabel('Fase do Dia', fontsize=11)
plt.ylabel('Total de Vítimas', fontsize=12)
plt.yticks([])

plt.savefig('06_sinistros_fase_dia.png', bbox_inches='tight')
plt.show()
```

```python
# Criação do gráfico de Frequência de Acidentes por dia da Semana

# Cria um gráfico em branco
plt.figure()

# Selecionando as variáveis do DF
dias_ordem = ['segunda-feira', 'terça-feira', 'quarta-feira', 'quinta-feira', 'sexta-feira', 'sábado', 'domingo']

# Calcular a contagem para que o Seaborn possa mapear as cores pela intensidade
contagem_dias = df['dia_semana'].value_counts().reindex(dias_ordem)

# Criando o gráfico com x e y
ax = sns.barplot(x=contagem_dias.index, y=contagem_dias.values, palette='Reds', hue=contagem_dias.values, legend=False)

# Adicionando rótulos
for container in ax.containers:
    ax.bar_label(container, padding=3, fontsize=12, fontweight='bold', color="black")

# Customizando títulos, fontes, tamanhos, margens, etc
plt.title('Frequência de Acidentes por Dia da Semana - PRF 2025', fontsize=14, fontweight='bold', color="black", pad=20)
plt.xticks(fontsize=12, fontweight='bold', rotation=0)
plt.xlabel('Fase do Dia', fontsize=11)
plt.yticks([])
plt.ylabel('Quantidade de Acidentes', fontsize=12)

plt.savefig('07_acidentes_dia_semana.png', bbox_inches='tight')
plt.show()
```

```python
# Criação do gráfico de Correlação

# Cria um gráfico em branco
plt.figure()

# Selecionando as variáveis do DF
corr = df[['pessoas','veiculos','mortos','feridos']].corr()

# Criando o gráfico
sns.heatmap(corr, annot=True, cmap='Reds')

# Customizando títulos, fontes, tamanhos, margens, etc
plt.title("Correlação entre variáveis: Pessoas x Veículos x Mortos x Feridos", fontsize=16, fontweight='bold', color="black", pad=15)
plt.xticks(fontsize=12, fontweight='bold')
plt.yticks([])

plt.savefig('08_correlacao.png', bbox_inches='tight')
plt.show()
```

```python
# Criação do gráfico de Outiliers

# Criando o gráfico
df[['mortos','feridos','pessoas']].plot(kind='box')

# Customizando títulos, fontes, tamanhos, margens, etc
plt.title("Detecção de outliers", fontsize=16, fontweight='bold', color="black", pad=15)
plt.xticks(fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')

plt.savefig('09_outliers.png', bbox_inches='tight')
plt.show()
```

```python
# Criação do gráfico de acidentes por mês

# Cria um gráfico em branco
plt.figure()

# Criando a variável mês
df['mes'] = df['data_inversa'].dt.month

# Selecionando as variáveis do DF
contagem_mes = df['mes'].value_counts().sort_index()

# Criando o gráfico
ax = sns.barplot(x=contagem_mes.index, y=contagem_mes.values, palette='Reds', hue=contagem_mes.values, legend=False)

# Adicionando rótulos
for container in ax.containers:
    ax.bar_label(container, padding=3, fontsize=12, fontweight='bold', color="black")

# Customizando títulos, fontes, tamanhos, margens, etc
plt.title("Acidentes por mês - PRF 2025", fontsize=16, fontweight='bold', color="black", pad=15)
plt.xlabel("Mês", fontsize=12)
plt.xticks(fontsize=12, fontweight='bold', rotation=0)
plt.ylabel("Quantidade", fontsize=12)
plt.yticks([])

plt.savefig('10_acidentes_mes.png', bbox_inches='tight')
plt.show()
```

```python
# Criação do gráfico de acidentes por hora do dia

# Cria um gráfico em branco
plt.figure()

# Criando a variável hora
df['hora_cheia'] = pd.to_datetime(df['horario'], format='mixed').dt.hour

# Selecionando as variáveis do DF
contagem_hora = df['hora_cheia'].value_counts().sort_index()

# Criando o gráfico
ax = sns.barplot(x=contagem_hora.index, y=contagem_hora.values, palette='Reds', hue=contagem_hora.values, legend=False)

# Adicionando rótulos
for container in ax.containers:
    ax.bar_label(container, padding=3, fontsize=12, fontweight='bold', color="black")

# Customizando títulos, fontes, tamanhos, margens, etc
plt.title("Acidentes por hora do dia - PRF 2025", fontsize=16, fontweight='bold', color="black", pad=15)
plt.xlabel("Hora", fontsize=12)
plt.xticks(fontsize=12, fontweight='bold', rotation=0)
plt.ylabel("Quantidade", fontsize=12)
plt.yticks([])

plt.savefig('11_acidentes_hora.png', bbox_inches='tight')
plt.show()
```
